```python
# !pip install pandas matplotlib seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./log.csv", names=["Threads", "N", "M", "Time (ms)"])
df
```

```python
sns.set_theme()

#     
unique_elements = df['N'].unique()
# unique_elements

for num_elements in unique_elements:
    subset_df = df[df['N'] == num_elements]
    plt.figure(figsize=(10, 6))
    plt.plot(subset_df['Threads'], subset_df['Time (ms)'], marker='o')
    plt.title(f'       = {num_elements**2}')
    plt.xlabel('Threads')
    plt.ylabel('Time (ms)')
    plt.xticks(subset_df['Threads'].unique())
    plt.grid(True)
    plt.show()
```
